```python
bins=(2,6.5,8)
```

```python
group_names=['bad','good']
df['quality']=pd.cut(df['quality'],bins=bins,labels=group_names)
```

good and bad are not asigned a numeric label.So we are using label encoder

```python
from sklearn.preprocessing import LabelEncoder
label_quality=LabelEncoder()
```

```python
df['quality']=label_quality.fit_transform(df['quality'])
```

```python
df.head()
```

```python
X=df.drop('quality',axis=1)
y=df[['quality']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix, classification_report
def models(X_train,y_train):
    #using decisiontreeclassifier
    from sklearn.tree import DecisionTreeClassifier
    tree=DecisionTreeClassifier(criterion='entropy',random_state=42)
    tree.fit(X_train,y_train)
    #using Randomforest
    
    from sklearn.ensemble import RandomForestClassifier
    forest=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=42)
    forest.fit(X_train,y_train)
    f_pred=forest.predict(X_test)
    print('Decision tree classifier training accuracy:',tree.score(X_train,y_train))
    print('Random Forest classifier training accuracy:',forest.score(X_train,y_train))
    return tree,forest   
```

```python
model=models(X_train,y_train)
```

```python
y_pred=0
for i in range(len(model)):
    y_pred=model[i].predict(X_test)
    cm=confusion_matrix(y_test,y_pred)
    
    TN,FP,FN,TP=confusion_matrix(y_test,model[i].predict(X_test)).ravel()
    print(cm)
    print('Testing accuracy={}'.format((TP+TN)/(TP+TN+FP+FN)))
    
```
